```python
import random

class RelationGenerator:
    
    def __init__(self, d=5, c=5):
        
        self._MATRIX_SIZE = 100

        self._matrix = [[(y,x) for x in range(1,self._MATRIX_SIZE+1)] for y in range(1,self._MATRIX_SIZE+1)] 
        self._domain = FiniteSet(*[i for i in range(1,d+1)])
        self._codomain = FiniteSet(*[i for i in range(1,c+1)])
        
    def get_domain(self):
        return self._domain
    
    def set_domain(self, size):
        self._domain = FiniteSet(*[i for i in range(1,size+1)])
    
    def get_codomain(self):
        return self._codomain
    
    def set_codomain(self, size):
        self._codomain = FiniteSet(*[i for i in range(1,size+1)])
        
    def get_matrix(self):
        return self._matrix
        
    def generate_binary_relation(self, count=10):
        
        matrix = self.get_matrix()
        
        for _ in range(count):
            
            relation = set()
        
            for i in range(len(self.get_domain())):
                for j in range(len(self.get_codomain())):
                    if random.randrange(2) == 0:
                        pass
                    else:
                        relation.add(matrix[i][j])
                    
            yield BinaryRelation(FiniteSet(*relation), self.get_domain(), self.get_codomain(), True)
                        
                        
    def generate_homogeneous_relation(self, count=10):
        
        matrix = self.get_matrix()
        
        for _ in range(count):
            
            relation = set()
        
            for i in range(len(self.get_domain())):
                for j in range(len(self.get_codomain())):
                    if random.randrange(2) == 0:
                        pass
                    else:
                        relation.add(matrix[i][j])
                    
            yield HomogeneousRelation(FiniteSet(*relation), self.get_domain(), True)          
        
    def generate_partial_order(self, count=10):
        
        matrix = self.get_matrix()
        
        for _ in range(count):
            
            relation = set()

            dict_in_relation = dict()
            dict_out_relation = dict()

            for p in self.get_domain():
                dict_in_relation[p] = set()

            for p in self.get_codomain():
                dict_out_relation[p] = set()
        
            for i in range(len(self.get_domain())):
                for j in range(len(self.get_codomain())):
                    if i == j: #reflexive
                        relation.add(matrix[i][j])
                    else:
                        if random.randrange(2) == 0:
                            pass                          
                        else:
                            if matrix[j][i] not in relation: #antisymmetric
                                
                                tmp_relation = relation.copy()
                                tmp_dict_in_relation = dict_in_relation.copy()
                                tmp_dict_out_relation = dict_out_relation.copy()
                                
                                tmp_relation.add(matrix[i][j])
                                tmp_dict_in_relation[i+1].add(j+1)
                                tmp_dict_out_relation[j+1].add(i+1)
                                
                                revoke = False
                                
                                start_array = [i+1]
                                stop_array = [j+1]
                                
                                while start_array != [] and (revoke == False): #transitive
           
                                    start = start_array.pop()
                                    stop = stop_array.pop()
                
                                    for p in tmp_dict_in_relation[stop]:
                                        if (p, start) not in tmp_relation:

                                            tmp_relation.add((start, p))
                                            tmp_dict_in_relation[start].add(p)
                                            tmp_dict_out_relation[p].add(start)

                                            start_array.append(start)
                                            stop_array.append(p)

                                        else:
                                            revoke = True
                                            break
                                            
                                    for p in tmp_dict_out_relation[start]:
                                        if (stop, p) not in tmp_relation:
        
                                            tmp_relation.add((p, stop))
                                            tmp_dict_in_relation[p].add(stop)
                                            tmp_dict_out_relation[stop].add(p)

                                            start_array.append(p)
                                            stop_array.append(stop)
                   
                                        else:
                                            revoke = True
                                            break

                                if revoke == True:
                                    pass
                                else:
                                    relation = tmp_relation.copy()
                                    dict_in_relation = tmp_dict_in_relation.copy()
                                    dict_out_relation = tmp_dict_out_relation.copy()
                            else:
                                pass
                                    
            yield PartialOrder(relation, self.get_domain(), True)
```
